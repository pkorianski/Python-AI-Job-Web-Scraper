```python

import urllib.request, urllib.parse, urllib.error
import bs4
from bs4 import BeautifulSoup
import requests
import re
import os
import pandas as pd
```

```python
##################
# INITIALIZATION #
##################

dict = []
jobs = []
companies = []
locations = []
salaries = []
summaries = []
```

```python
##################
#    METHODS     #
##################

def insert_jobreq():
    l = "https://www.indeed.com/q-python-l-Boston,-MA-jobs.html"
    page = requests.get(l)
    soup = BeautifulSoup(page.text, "html.parser")
    #print(soup.prettify())

    # Inserting job titles into the job list
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
            jobs.append(a["title"])
    #print(jobs)

    # Inserting companies
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        company = div.find_all(name="span", attrs={"class":"company"})
        if len(company) > 0:
            for b in company:
                companies.append(b.text.strip())
                break
        else:
            sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
            for span in sec_try:
                companies.append(span.text.strip())
    #print(companies)

    # Insert locations
    spans = soup.findAll('span', attrs={'class': 'location'})
    for span in spans:
        locations.append(span.text)
    #print(locations)

    # Insert salaries
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        try:
            salaries.append(div.find('nobr').text)
        except:
            try:
                div_two = div.find(name="div", attrs={"class":"sjcl"})
                div_three = div_two.find("div")
                salaries.append(div_three.text.strip())
            except:
                salaries.append("Nothing_found")
    #print(salaries)

    # Job Description
    spans = soup.findAll('span', attrs={'class': 'summary'})
    for span in spans:
        summaries.append(span.text.strip())
    print(summaries)

    max_results_per_city = 100
    city_set = ['New+York','Chicago','San+Francisco', 'Austin', 'Seattle', 'Los+Angeles', 'Philadelphia', 'Atlanta', 'Dallas', 'Pittsburgh', 'Portland', 'Phoenix', 'Denver', 'Houston', 'Miami', 'Washington+DC', 'Boulder']
    columns = ["city", "job_title", "company_name", "location", "summary", "salary"]
    sample_df = pd.DataFrame(columns = columns)

    #ui = input("Press 1 to go to main menu: ")
```
